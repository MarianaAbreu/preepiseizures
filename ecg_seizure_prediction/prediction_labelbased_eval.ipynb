## Evaluation of the HRV Seizure Prediction 

```python
import json
import os

import pandas as pd

from preepiseizures.src import Patient
```

```python
patients_json = json.load(open('patient_info.json'))

patient = 'YWJN'
patient_info = Patient.patient_class(patient)
patient_info.get_seizure_annotations()
patient_info.seizure_table

failed_file = os.path.join(f'data{os.sep}{patient}_prediction_eval_labelbased_failed.txt')
results_file = os.path.join(f'data{os.sep}{patient}_prediction_eval_labelbased.csv')
```

```python
if os.path.exists(failed_file):
    print(f'Patient {patient} failed')
    with open(failed_file, 'r') as f:
        failed = f.read()
    print(failed)
else:
    print(f'Patient {patient} succeeded')
```

```python
results_data = pd.read_csv(results_file)
print(results_data.iloc[0])
```

```python
for patient in patients_json.keys():

    failed_file = os.path.join(f'data{os.sep}{patient}_prediction_eval_labelbased_failed.txt')
    results_file = os.path.join(f'data{os.sep}{patient}_prediction_eval_labelbased.csv')
    if os.path.exists(failed_file):
        print(f'Patient {patient} failed')
        with open(failed_file, 'r') as f:
            failed = f.read()
        print(failed)
        continue
    else:
        print(f'Patient {patient} succeeded')
        results_data = pd.read_csv(results_file)
    print(results_data.iloc[0])
        
```
